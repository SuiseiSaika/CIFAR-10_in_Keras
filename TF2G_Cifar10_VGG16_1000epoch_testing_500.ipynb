```
from tensorflow.keras import *
from google.colab import files
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```
def sdlevelsmall(bn, sd, lastsd = False):
  initializer = initializers.he_normal(seed=None)
  l2r = regularizers.l2(0.001)

  model = models.Sequential()
  model.add(layers.Conv2D(64, (3, 3), padding='same', use_bias='true', kernel_initializer=initializer, kernel_regularizer=l2r ,input_shape=(32, 32, 3)))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.125))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(64, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.125))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(128, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.125))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(128, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.125))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.25))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.25))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.25))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.5))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.5))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.5))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.5))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.5))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.5))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.GlobalAveragePooling2D())

  model.add(layers.Dense(10, activation='softmax'))
  return model
```

```
def sdlevellarge(bn, sd, lastsd = False):
  initializer = initializers.he_normal(seed=None)
  l2r = regularizers.l2(0.001)

  model = models.Sequential()
  model.add(layers.Conv2D(64, (3, 3), padding='same', use_bias='true', kernel_initializer=initializer, kernel_regularizer=l2r ,input_shape=(32, 32, 3)))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.3))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(64, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.3))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(128, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(128, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(256, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.Conv2D(512, (3, 3), padding='same', kernel_initializer=initializer, kernel_regularizer=l2r , use_bias='true'))
  if bn:
    model.add(layers.BatchNormalization())
  if sd:
   model.add(layers.SpatialDropout2D(0.4))
  model.add(layers.Activation('elu'))
  model.add(layers.MaxPooling2D((2, 2)))

  model.add(layers.GlobalAveragePooling2D())

  model.add(layers.Dense(10, activation='softmax'))
  return model
```

```
modelnamelist = [ 'SDunfixed-small',
                  'SDunfixed-large']              

modellist = [ sdlevelsmall(bn = True, sd = True),
              sdlevellarge(bn = True, sd = True)]
```

```
Adam = optimizers.Adam(lr=1e-4, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0)
batch = 128
epoch = 500
history = []
for m in modellist:
    m.compile(optimizer= Adam ,
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

    aug = preprocessing.image.ImageDataGenerator(
                                                # zca_whitening=True,
                                                rotation_range=25,
                                                width_shift_range=0.1,
                                                height_shift_range=0.1,
                                                horizontal_flip=True)

    aug.fit(train_images)
    # early_stop = callbacks.EarlyStopping(monitor='loss', patience=10, verbose=2)

    h =  m.fit_generator(
                                generator=aug.flow(train_images, train_labels, batch_size=batch),
                                epochs=epoch, 
                                steps_per_epoch=int(np.ceil(train_images.shape[0] / batch)),
                                verbose = 2,
                                validation_data=(test_images,  test_labels),
                                # callbacks=[early_stop]
                         )
    history.append(h)
```

```
for i in range(len(modelnamelist)):
  print(modelnamelist[i])
  test_loss, test_acc = modellist[i].evaluate(test_images,  test_labels, verbose=2)
```

```
for i in range(len(history)):
  plt.plot(history[i].history['acc'], label='accuracy of ' + modelnamelist[i])
  plt.plot(history[i].history['val_acc'], label = 'val_accuracy of ' + modelnamelist[i])
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.savefig("adam_acc_regularizer_2.png", dpi=500)
files.download("adam_acc_regularizer_2.png") 
```
